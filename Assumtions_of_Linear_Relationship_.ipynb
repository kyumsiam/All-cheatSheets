```python
df.head()
```

```python
X = df.iloc[:,0:3].values
y = df.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()

model.fit(X_train,y_train)
```

```python
# Residual
y_pred = model.predict(X_test)
residual = y_test - y_pred
```

# 2. Multicollinearity

Why multicollinearity Hampers the dataset ??
Ans : suppose Two researchers from two backgrounds physics and chemistry so it's easy to predict how much impact of each person on the work . If they belong from same background physics it's actually hard to do . So needs to remove the Multicolinearity .

```python
X_train.shape
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = []

for i in range(X_train.shape[1]):
    vif.append(variance_inflation_factor(X_train, i))
```

```python
pd.DataFrame({'vif': vif}, index=df.columns[0:3]).T
```

**IF VIF value 5 or more than 5 it's called Multicolinearity**
